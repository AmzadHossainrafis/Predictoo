# Shapiro-Wilk test

```python
stat, p = shapiro(Fundamentals_df)
print('Test statistic:', stat)
print('p-value:', p)
if p > 0.05:
    print('Data is normally distributed')
else:
    print('Data is not normally distributed')
```

# MinMaxScaler

```python
def min_max_scaling(col):
    return (col - col.min()) / (col.max() - col.min())
```

# Correlation Analysis

## Pearson

### Weighted corr

```python
from scipy.stats import pearsonr
import numpy as np
```

```python
Final_df['GIB']
```

```python
Final_df = Final_df.fillna(0)
```

```python
Final_df = Final_df.replace([np.inf, -np.inf], 1e10)
```

```python
weight = [0.65] * 25  + [0.95]*1 + [0.3]*400
```

```python
weighted_corrs = [w * pearsonr(Final_df['1JANATAMF'], Final_df['AAMRATECH'])[0] for w in weight]
corr = sum(weighted_corrs) / sum(weight)
corr
```

```python
weights = [0.65] * 25 + [0.90] + [0.30] * 400
```

```python
corr_df = Final_df.iloc[:, 0:20].corr(method='pearson', min_periods=1)
```

```python
weighted_corr_matrix = pd.DataFrame(index=corr_df.columns, columns=corr_df.columns)
for col1 in corr_df.columns:
    for col2 in corr_df.columns:
        weighted_corrs = [w * pearsonr(Final_df[col1], Final_df[col2])[0] for w in weight]
        corr = sum(weighted_corrs) / sum(weight)
        weighted_corr_matrix.loc[col1, col2] = corr
```

```python
weighted_corr_matrix
```
