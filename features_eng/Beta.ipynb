```python
path = "/content/drive/MyDrive/Stocknow"
```

```python
data_frames = []
for file in os.listdir(path):
    if file.endswith(".csv"):
        file_path = os.path.join(path, file)
        df = pd.read_csv(file_path)
        data_frames.append(df)
```

```python
result_df = pd.concat(data_frames, axis=0, ignore_index=True)
```

```python
result_df
```

```python
result_df['Year'] = result_df['Date'].astype(str).str[:4].astype(int)
```

```python
result_df['Month'] = result_df['Date'].astype(str).str[4:6].astype(int)
```

```python
result_df['Day'] = result_df['Date'].astype(str).str[6:8].astype(int)
```

```python
result_df['Date(R)'] = pd.to_datetime(result_df[['Year', 'Month', 'Day']])
```

```python
filtered_df_DSEX = result_df[result_df["Scrip"] == "DSEX"]
```

```python
filtered_df_DSEX = filtered_df_DSEX.sort_values(by="Date(R)", ascending=False)
```

```python
filtered_df_DSEX
```

```python
filtered_df_DSEX['Daily Return(DSEX)'] = filtered_df_DSEX['Close'].pct_change()
```

```python
filtered_df_DSEX
```

```python
filtered_df_DSEX['Variance'] = filtered_df_DSEX['Daily Return(DSEX)'].var()
filtered_df_DSEX
```

```python
filtered_df_SQURPH = result_df[result_df["Scrip"] == "SQURPHARMA"]
```

```python
filtered_df_SQURPH = filtered_df_SQURPH.sort_values(by="Date(R)", ascending=False)
```

```python
filtered_df_SQURPH
```

```python
filtered_df_SQURPH_M = pd.merge(filtered_df_SQURPH, filtered_df_DSEX[['Date(R)','Daily Return(DSEX)','Variance',]], on='Date(R)', how='inner')
```

```python
filtered_df_SQURPH_M
```

```python
filtered_df_SQURPH_M['Daily Return_SQ'] = filtered_df_SQURPH_M['Close'].pct_change()
```

```python
filtered_df_SQURPH_M.head()
```

```python
filtered_df_SQURPH_M['Covariance'] = filtered_df_SQURPH_M['Daily Return_SQ'].cov(filtered_df_SQURPH_M['Daily Return(DSEX)'])
```

```python
filtered_df_SQURPH_M
```

```python
filtered_df_SQURPH_M['Beta'] = filtered_df_SQURPH_M['Covariance']/filtered_df_SQURPH_M['Variance']
```
