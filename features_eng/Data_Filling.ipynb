## Pivot point

```python
new_df['Pivot Point'] = (new_df['High']+new_df['Low']+new_df['Close'])/3
```

```python
new_df.head(3)
```

## Resistance 1&2

```python
new_df['R1'] = (new_df['Pivot Point']*2)-new_df['Low']
```

```python
new_df['R2'] = new_df['Pivot Point'] + (new_df['High']-new_df['Low'])
```

## Support 1&2

```python
new_df['1st Support'] = (new_df['Pivot Point']*2) - new_df['High']
```

```python
new_df['2nd Support'] = new_df['Pivot Point'] - (new_df['High']-new_df['Low'])
```
